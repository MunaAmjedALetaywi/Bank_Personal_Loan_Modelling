```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
y = dataset['Personal Loan']
X = dataset.drop(['Personal Loan'], axis=1).copy()
X_train, X_valid, y_train, y_valid = train_test_split(X,y, test_size=0.3, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
mess = []
neighbors = []
def knn_model(n):
    KNN_model = KNeighborsClassifier(n_neighbors=n)
    KNN_model.fit(X_train,y_train)
    score_train = KNN_model.score(X_train,y_train)
    score_test = KNN_model.score(X_valid,y_valid)
    return score_test

for n in np.arange(1,30):
    if n%2 !=0:
        neighbors.append(n)
        test_score_val = knn_model(n)
        mess.append(1-test_score_val)
```

```python
best_neighbors_val = neighbors[mess.index(min(mess))]
print('Best Neighbor is', best_neighbors_val)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
KNN_model = KNeighborsClassifier(n_neighbors=n)
KNN_model.fit(X_train,y_train)
clf = KNN_model.fit(X_train,y_train)
prediction = clf.predict(X_valid)
# Create confusion matrix
results=confusion_matrix(y_valid,prediction)
# Display accuracy score
print(accuracy_score(y_valid, prediction))
print("Confusion Matrix :")
print(results)
print("Report :")
print(classification_report(y_valid,prediction))
```

```python
#we have noise in Income
df1 = dataset[dataset['Income']>160].index
dataset = dataset.drop(df1)
sns.boxplot(x='Income' , data=dataset)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
x = dataset.drop('Personal Loan' , axis=1)
#Target
y = dataset['Personal Loan']
#Prepare the train data and test data and the size for testing
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.3 , random_state=0)
print('x_train shape : ' , x_train.shape)
print('x_test shape : ' , x_test.shape)
print('y_train shape : ' , y_train.shape)
print('y_test shape : ' , y_test.shape)
Scaler = StandardScaler()
x_train_scaled = Scaler.fit_transform(x_train)
x_test_scaled = Scaler.transform(x_test)
LR = LogisticRegression()
LR.fit(x_train_scaled,y_train)
y_pred = LR.predict(x_test_scaled)
LR_score = metrics.accuracy_score(y_test,y_pred)
print('accuracy_score : ' , LR_score)
result = confusion_matrix(y_test,y_pred)
print("Confusion Matrix :")
print(result)
```

```python
from sklearn.naive_bayes import GaussianNB
x_train1,x_test1,y_train1,y_test1 = train_test_split(x,y,test_size=0.3,random_state=10)
nb = GaussianNB()
nb.fit(x_train1,y_train1)
y_pred = nb.predict(x_test1)
print(accuracy_score(y_test1,y_pred))
print("-------------------------------------------------------")
results=confusion_matrix(y_test1,y_pred)
print("Confusion Matrix :")
print(results)
print("Report :")
print(classification_report(y_test1,y_pred))
```
